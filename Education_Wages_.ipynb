## **Summary Statistics**
--------------------------

- Mean, Median, Mode, Standard Deviation, and Interquartile Range for Bachelor's Degree

```python
import numpy as np
from scipy.stats import mode
from scipy.stats import iqr


# Sample data
data = df['bachelors_degree']

# Calculating mean, median, mode, standard deviation, and interquartile range
mean_value = np.mean(data)
median_value = np.median(data)
mode_value = mode(data, keepdims = False)
std_deviation = np.std(data, ddof=1)  # Using ddof=1 for sample standard deviation
interquartile_range = iqr(data)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value.mode)
print("Standard Deviation:", std_deviation)
print("Interquartile Range:", interquartile_range)
```

## **OLS Regression Results / Models**
--------------------------


```python
independent_vars = ['white_less_than_hs', 'white_high_school', 'white_some_college', 'white_bachelors_degree', 'white_advanced_degree',
                     'black_less_than_hs', 'black_high_school', 'black_some_college', 'black_bachelors_degree', 'black_advanced_degree',
                     'hispanic_less_than_hs', 'hispanic_high_school', 'hispanic_some_college', 'hispanic_bachelors_degree', 'hispanic_advanced_degree']

dependent_var = 'some_college'

X = df[independent_vars]

X = sm.add_constant(X)

model = sm.OLS(df[dependent_var], X).fit()

print(model.summary())
```

```python
residuals = model.resid
plt.scatter(df[dependent_var], residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

**Heatmap:** Coorelation between Bachelors Degree Earnings within each Demographic

```python
df1=df[df['year']>1999]
#print(df1)
df2 = pd.DataFrame()
df2['black_bachelors_degree']=df['black_bachelors_degree']
df2['white_bachelors_degree']=df['white_bachelors_degree']
df2['hispanic_bachelors_degree']=df['hispanic_bachelors_degree']
df2['bachelors_degree']=df['bachelors_degree']

data_corr = df2.copy()
correlation = data_corr.corr()
sb.heatmap(correlation,fmt='.2f', cmap = 'coolwarm',vmin=-1,vmax=1)
plt.show()
print(correlation)
```

- This heatmap above displays direct coorelation between the wages of each demographic (White, Black, and Hispanic). There is high coorelation among the overall wage distribution.

**Box Plot:** Bachelors Degree Earnings per Demographic

```python
columns_of_interest = ['bachelors_degree', 'white_bachelors_degree', 'hispanic_bachelors_degree', 'black_bachelors_degree']
data_for_plot = df[columns_of_interest].melt(var_name='Group', value_name='Earnings')
# Define the color for each group
color_palette = {
    "bachelors_degree": "blue", 
    "white_bachelors_degree": "green", "hispanic_bachelors_degree": "orange", "black_bachelors_degree": "red"      
}

# Create the boxplot with the specified color palette
plt.figure(figsize=(10, 6))
sb.boxplot(x='Group', y='Earnings', data=data_for_plot, palette = color_palette)
plt.title('Box Plot: Earnings for Ethnic Groups with Bachelors Degree')
plt.xlabel('Ethnic Group')
plt.ylabel('Hourly Wage')
plt.show()
```

--------------------------
**Line Graph:** Average Wages Over Years (2000-2022) by Demographic

```python
demographic_columns = ['year', 'bachelors_degree','white_bachelors_degree', 'black_bachelors_degree', 'hispanic_bachelors_degree']
df_filtered = df[df['year'].between(2000, 2022)][demographic_columns]

color_palette = {
    "bachelors_degree": "blue", 
    "white_bachelors_degree": "green", "hispanic_bachelors_degree": "orange", "black_bachelors_degree": "red"      
}

# Melting the DataFrame for easier plotting with seaborn
df_melted = df_filtered.melt(id_vars='year', var_name='Demographic', value_name='AverageWage')

# Generate the line graph
plt.figure(figsize=(12, 8))
sb.lineplot(data=df_melted, x='year', y='AverageWage', hue='Demographic', palette = color_palette)
plt.title('Line Graph: Average Wages Over Years (2000-2022) by Demographic')
plt.xlabel('Year')
plt.ylabel('Average Wage')
plt.xticks(rotation=45)
plt.show()
```

--------------
**Bar Chart:** Average Wages Over Years (2000-2022) by Demographic

```python
# Select relevant columns for the bar chart (bachelor's degree holders among different demographics)
bar_chart_columns = ['year', 'white_bachelors_degree', 'hispanic_bachelors_degree', 'black_bachelors_degree']
bar_chart_df = df[df['year'].between(2000, 2022)][bar_chart_columns]

# Reshaping the data for the bar chart
bar_chart_data = bar_chart_df.melt(id_vars='year', var_name='Demographic', value_name='AverageWage')

green_palette = sb.color_palette("Greens", n_colors=3)

# Generating the Bar Chart
plt.figure(figsize=(12, 8))
sb.barplot(data=bar_chart_data, x='year', y='AverageWage', hue='Demographic', palette = green_palette)
plt.title('Bar Chart: Average Wages Over Years (2000-2022) by Demographic')
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Average Wage')
plt.show()
```
